# **Data Preprocessing for Model**

```python
X_train = data.iloc[:, 2:32].values
y_train = data.iloc[:, 1].values
X_train, X_test, y_train, y_test = train_test_split(X_train, y_train, test_size=0.2, random_state=0)

print('Train dataset shape:',X_train.shape)
print('Test dataset shape', y_train.shape)
```

```python
st  = StandardScaler()
```

```python
X_train  = st.fit_transform(X_train)
X_test  = st.fit_transform(X_test)
```

# **Checking Which model gives good score Linear Regression or Logistic Regretion**

# **Linear Regression**

```python
# Linear Regression
log = LinearRegression()
log.fit(X_train, y_train)
```

```python
log.score(X_train, y_train)
```

# **Logistic Regression**

```python
# Logistic Regression
log1 = LogisticRegression()
log1.fit(X_train, y_train)
```

```python
log1.score(X_train, y_train)
```

```python
accuracy_score(y_test, log1.predict(X_test))
```

```python
print(classification_report(y_test, log1.predict(X_test)))
```

# **Saving the model in pickle for later use**

```python
import pickle
pickle.dump(log1, open("model.pkl", "wb"))
pickle.load(open("model.pkl", "rb"))
```
